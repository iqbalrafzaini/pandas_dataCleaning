###DUPLICATE

```python
data_dict = {'name':['Abdul', 'Ahmad', 'Jose', 'Bianca', 'Rahayu', 'Jermaine', 'Ahmad', 'Jose'],
            'age':[25, 23, 18, 45, 33, 15, 23, 18],
            'hobby':['sport', 'sport', 'art', 'sport', 'culinary', 'academics', 'sport', 'culinary']}

df_dup = pd.DataFrame(data_dict)
df_dup
```

```python
df_dup.duplicated(keep=False) #jika yg keluar true maka terdapat duplicate
```

```python
df_dup[df_dup.duplicated(subset=['name','age'],keep=False)]
```

```python
df_dup.drop_duplicates()
```
